# Updating Notebook Banners

```python
import os
import json
import pandas as pd
```

## Functions to Replace or Add Banners

```python
#notebook_name_and_path = './TechnologyStudies/IntroductionToPython/introduction-to-python.ipynb'

first_cell_data = {'cell_type': 'markdown','metadata': {},'source': ['![Callysto.ca Banner](https://github.com/callysto/curriculum-notebooks/blob/master/callysto-notebook-banner-top.jpg?raw=true)']}
last_cell_data = {'cell_type': 'markdown','metadata': {},'source': ['[![Callysto.ca License](https://github.com/callysto/curriculum-notebooks/blob/master/callysto-notebook-banner-bottom.jpg?raw=true)](https://github.com/callysto/curriculum-notebooks/blob/master/LICENSE.md)']}

def replace_first_cell(notebook_name_and_path):
    original_file = open(notebook_name_and_path, 'r')
    notebook_contents = json.load(original_file)
    original_file.close()    
    del notebook_contents['cells'][0]
    notebook_contents['cells'].insert(0, first_cell_data)
    with open(notebook_name_and_path, 'w') as notebook_file:
        json.dump(notebook_contents, notebook_file)

def replace_last_cell(notebook_name_and_path):
    original_file = open(notebook_name_and_path, 'r')
    notebook_contents = json.load(original_file)
    original_file.close()
    del notebook_contents['cells'][-1]
    notebook_contents['cells'].append(last_cell_data)
    with open(notebook_name_and_path, 'w') as notebook_file:
        json.dump(notebook_contents, notebook_file)

def add_first_and_last_cells(notebook_name_and_path):
    original_file = open(notebook_name_and_path, 'r')
    notebook_contents = json.load(original_file)
    original_file.close()
    notebook_contents['cells'].insert(0, first_cell_data)
    notebook_contents['cells'].append(last_cell_data)
    #import pprint
    #pprint.pprint(notebook_contents) #['cells'][0]['source'])
    #with open(notebook_name.replace('.ipynb', '-test.ipynb'), 'w') as notebook_file:
    with open(notebook_name_and_path, 'w') as notebook_file:
        json.dump(notebook_contents, notebook_file)
```

## Create First and Last Cells DataFrame

```python
df = pd.DataFrame(columns=['Notebook', 'Path', 'First Cell', 'Last Cell'])
for root, dirs, files in os.walk("."):
    for filename in files:
        if filename.endswith('.ipynb'):
            if not 'checkpoint' in filename:
                notebook_name = filename[:-6]
                file = os.path.join(root, filename)
                notebook = json.load(open(file))
                first_cell = notebook['cells'][0]['source']#[0]
                last_cell = notebook['cells'][-1]['source']#[0]
                df = df.append({'Notebook':notebook_name, 'Path':root, 'First Cell':first_cell, 'Last Cell':last_cell}, ignore_index=True)
# create length of cell columns
df['First Cell Length'] = df['First Cell'].str.len()
df['Last Cell Length'] = df['Last Cell'].str.len()
df.tail()
```

```python
# if oops then remove those columns
df = df.drop('First Cell Length', axis=1).drop('Last Cell Length', axis=1)
```
